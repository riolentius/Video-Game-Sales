```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as clr
import numpy as np
from matplotlib.widgets import Slider,Button,RadioButtons

# read file of the csv
genre_region = pd.read_csv('/Users/ice/Pictures/Rio/Dataset/Kaggle/Video Game Genre/genre_region_totals.csv')
genre_total = pd.read_csv('/Users/ice/Pictures/Rio/Dataset/Kaggle/Video Game Genre/genre_totals.csv')
year_genre = pd.read_csv('/Users/ice/Pictures/Rio/Dataset/Kaggle/Video Game Genre/year_genre_region.csv')

print(year_genre.head(10))
```

```python
# checking data type of the csv
data_genre_region = genre_region.dtypes
data_genre_total = genre_total.dtypes
data_year_genre = year_genre.dtypes

print(data_year_genre)
```

```python
# creating visualization for genre total and change data types

# visualization with pie chart for the genre total
labels = genre_total['genre']
sizes = genre_total['sales']
explode = np.zeros(len(sizes))
explode[sizes.argmax()] = 0.1

fig1, ax1 = plt.subplots(figsize=(20,10))
fig1.subplots_adjust(0.3,0,1,1)

ax1.pie(sizes, autopct='%1.1f%%',explode=explode,shadow=True,startangle=90)
total = sum(sizes)
plt.legend(loc='upper left',labels=labels)
ax1.axis('equal')
plt.title("Video Game Sales by Genre From 1980 - 2020")

plt.show()
```

```python
# visualization for Year Genre for worldwide in  for simulation genre

region_year = year_genre[(year_genre["release_year"].between(2010,2020)) & (year_genre['region'] == 'Worldwide') & (year_genre['genre'] == 'Action')]

region_year.sort_index()

objects = region_year['release_year']
y_position = np.arange(len(objects))
sales = region_year['sales']


plt.barh(y_position, sales, align='center')
plt.yticks(y_position,objects)
plt.title('Video Games Sales By Genre Action for Worldwide from 2010 - 2020')
plt.show()
```
